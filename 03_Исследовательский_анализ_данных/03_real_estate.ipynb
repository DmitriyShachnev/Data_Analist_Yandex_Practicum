## Предобработка данных

### Устранение неявных дубликатов и заполнение пропусков

```python
# выведем количество пропущенных значений для каждого столбца
data.isna().sum()
```

```python
# заменим букву "ё" на "е" в столбе, содержащим наименования населенных пунктов
data['locality_name'] = data['locality_name'].replace('ё', 'е', regex=True)
```

```python
# для устранения неявных дубликатов в наименованиях населенных пунктов исключим из них часть слов
data['locality_name'] = data['locality_name'].replace(
    ['коттеджный ', 'станции ', 'садовое товарищество ', 
     'село ', 'городской поселок ', 'деревня ', 'поселок ', 'поселок городского типа ', 'городского типа ', 'садоводческое некоммерческое товарищество ', \
     'поселок при железнодорожной станции ', 'поселок при железнодорожной ', 'при железнодорожной '], '', regex=True)
```

```python
# выведем количество уникальных значений в стобце 'locality_name'
data['locality_name'].nunique()
```

Всего имеется 305 уникальных наименований населенных пунктов (без NaN).

```python
data['ceiling_height'].describe()
```

Максимальное значение высоты потолка равно 100 м. Вероятно, в данных имеются ошибки. Найдем значения, превышающие 10 м. и заменим их этими же значениями, деленными на 10.

```python
# найдем значения высоты потолка, превышающие 10 м., и заменим их этими же значениями, деленными на 10.
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height'] < 10, data['ceiling_height']/10)
```

```python
# заменим пропущенные значения в поле, содержащем информацию о высоте потолков, медианными значениями высоты потолков в квартирах,
# расположенных в тех же населенных пунктах, в которых находятся квартиры с пропущенными значениями высоты потолков. 
for i in data['locality_name'].unique():
    data.loc[(data['ceiling_height'].isna()), 'ceiling_height'] = data.loc[(data['locality_name'] == i), 'ceiling_height'].median()
```

```python
# заменим пропущенные данные в поле, содержащем информацию о наличии и количестве балконов в квартирах, нулевым значением
data['balcony'] = data['balcony'].fillna(0)
```

```python
# заполним пропущенные данные в поле, содержащем информацию о жилой площади квартир
# для этого рассчитаем коэффициент соотношения медианных значений жилой и общей площади квартир (где эти значения имеются),
# после чего умножим значение общей площади на данный коэффициент и подставим получившиеся значения вместо пропущенных.
area_koef = data[data['living_area'].isnull() == False]['living_area'].median() \
/ data[data['total_area'].isnull()== False]['total_area'].median() 
data['living_area'] = data['living_area'].fillna(data['total_area'] * area_koef)
```

```python
# аналогично заполним пропущенные данные в поле, содержащем информацию о площади кухни в квартирах
area_koef = data[data['kitchen_area'].isna() == False]['kitchen_area'].median() \
/ data[data['total_area'].isna()== False]['total_area'].median() 
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area'] * area_koef)
```

```python
# заменим пропущенные значения в поле, содержащем информацию о расстоянии до центра города, медианными значениями
for i in data['locality_name'].unique():
    data.loc[(data['city_centers_nearest'].isna()), 'city_centers_nearest'] = \
    data.loc[(data['locality_name'] == i), 'city_centers_nearest'].median()
```

```python
# заменим пропущенные значения в поле, содержащем информацию о количестве этажей в доме, где находится квартира:
for i in data['locality_name'].unique():
    data.loc[(data['floors_total'].isna()), 'floors_total'] = data.loc[(data['locality_name'] == i), 'floors_total'].median()    
```

```python
# преобразуем тип данных столбца 'first_day_exposition' из строчных значений в дату 
# выберем данные с пустыми значениями срока размещения объявлений и сгруппируем их по годам
# построим графикраспределения пропусков по годам размещения объявлений
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y.%m.%d')
data['first_day_exposition_year'] = data['first_day_exposition'].dt.year
data_exposition_gap = data[data['days_exposition'].isna()]
data_exposition_gap_grouped = data_exposition_gap.groupby('first_day_exposition_year')['first_day_exposition_year'].count()
data_exposition_gap_grouped.plot(kind='bar', label='Пропуски данных о сроках размещения объявлений по годам', grid=True, legend=True) \
.set(title='Пропуски данных о сроках размещения объявлений, по годам', xlabel='год подачи объявления.', ylabel='количество пропусков в данных') 
```

Как видно из графика, наибольшее количество пропусков в данных о продолжительности размещения объявлений о продаже наблюдается в отношении объявлений, впервые размещенных в 2019 году, наименьшее количество - в отношении объявлений, размещенных в 2014 году. С каждым последущим годом размещения объявлений о продаже увеличивается число пропусков. Это свидетельствует о постоянном росте незакрытых объявлений.  Можно предположить, что объявления не были  сняты с публикации после продажи квартиры, либо квартиры по ним еще не проданы.  
Будем считать, что квартиры были проданы и объявления сняты с публикации.

```python
# заменим пропущенные значения в поле, содержащем информацию о количестве дней размещения объявлений с момента размещения до снятия:
for i in data['locality_name'].unique():
    data.loc[data['days_exposition'].isna(), 'days_exposition'] = \
    data.loc[(data['locality_name'] == i), 'days_exposition'].median()
```

```python
# заменим пропущенные данные в поле, содержащем информацию о принадлежности квартиры к апартаментам, на булево значение False.
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Вывод: мы устранили неявные дубликаты в наименованиях населенных пунктов, заполнили пропущенные данные в полях, содержащих информацию о высоте потолков, наличии и количестве балконов в квартирах, жилой площади, площади кухни, расстоянии до центра города, количестве этажей в доме, где находится квартира, количестве дней размещения объявлений с момента размещения до снятия, принадлежности квартиры к апартаментам.  
Пропуски данных присутствуют также в следующих столбцах таблицы: locality_name, airports_nearest, parks_around_3000, parks_nearest, ponds_around_3000, ponds_nearest. Замена пропусков на средние или медианные значения в данном случае нецелесообразна, так как может повлиять на результаты исследования.    
Пропуски в данных о названии населенного пункта, количестве этажей в доме, в котором располагается квартира, жилой площади квартиры и площади кухни, вероятно, допущены самими пользователями и связаны с человеческим фактором. Также пропуски могут быть вызваны ошибками при выгрузке данных.

### Анализ зависимости стоимости объектов от расстояния до центра города в г. Санкт-Петербург

```python
# выделим квартиры в Санкт-Петербурге и рассчитаем среднюю цену 1 кв.м. в квартирах, расположенных на определенном расстоянии от центра города
data_SPb = data_filtered.query('locality_name == "Санкт-Петербург"')
data_SPb_pivot = data_SPb.pivot_table(index='city_centers_nearest_km', values='square_meter_price', aggfunc='mean').reset_index()
data_SPb_pivot.head(10)
```

```python
# построим график зависимости стоимости 1 кв.м. от расстояния дома до центра города
data_SPb_pivot.plot(x='city_centers_nearest_km', y='square_meter_price', kind='bar', grid=True ) \
.set(title='Зависимость цены 1 кв.м. объекта от расстояния до центра города', xlabel='расстояние до центра города, км.', ylabel='цена объекта, млн.руб.')  
```

```python
data_SPb_pivot.corr()
```

Между ценой за 1 кв.м. площади квартиры и расстоянием до центра города есть определенная зависимость.  
Зависимость отрицательная: при увеличении расстояния до центра города, стоимость 1 кв.м. площади квартиры уменьшается.  
Коэффициент корреляции составляет -0,76.  
Наибольший уровень цен за 1 кв.м. жилой площади наблюдается в квартирах, расположенных в домах на расстоянии 1-2, 4-8 км. Также имеется локальный пик на отметке 27 км., который объясняется наличием одного дорогого объекта с ценой 194 тыс.руб. за 1 кв.м., что существенно выше среднего уровня цен.
